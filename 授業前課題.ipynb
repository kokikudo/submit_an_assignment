# 43回紙を折った時の距離を算出

```
import time # 処理時間を計測するためにtimeライブラリをimport
```



> **べき乗で算出**



```
start = time.time()

THICKNESS = 0.00008
count = 43
folded_thickness = THICKNESS * (2**count)
print("厚さ： {:.2f}km".format(folded_thickness/1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

42回だと約35万kmとなり、月までの距離（約38万km）に届かないので、43回折る必要がある。



> **for文で算出** (使用できる算術演算子は四則演算（+、-、*、/）のみ)



```
start = time.time()

count_of_distance = 2

for _ in range(1, 43):
  count_of_distance  = count_of_distance * 2

folded_thickness = THICKNESS * count_of_distance

print("厚さ： {:.2f}km".format(folded_thickness/1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

処理時間を比較しやすくするため、繰り返し回数を増やしてみる

```
start = time.time()

count_of_distance = 2

# 繰り返し回数を10倍にしてみる
for _ in range(1, 43*10):
  count_of_distance  = count_of_distance * 2

folded_thickness = THICKNESS*count_of_distance

print("厚さ： {:.2f}km".format(folded_thickness/1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

繰り返し回数を増やした方が処理速度がはやい？

> **グラフを作成するため、繰り返し処理で出した値をリストに格納**



```
list_distance = []
for i in range(0, 44):
  list_distance.append(THICKNESS * (2**i))
len(list_distance)
```



> **グラフ作成**



```
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(list_distance)
plt.show()
```

40回目あたりから1億mを超え、そこから急激に増加している



> **折れ線グラフのカスタマイズ**



```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(list_distance, color='green')# 色を緑に
plt.show()
```

```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
# 点を▼、色を青、線を点線に変更
plt.plot(list_distance, marker='v', color='blue', linestyle=':')
plt.show()
```

```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
# 1億回目に補助線を追加
plt.hlines(1e+8, 0, len(list_distance), 'gray', linestyles='dashed')
plt.plot(list_distance, marker='v', color='blue', linestyle=':')
plt.show()
```

40回目までは1億を超えていない。41回目で初めて超えることがわかる。
